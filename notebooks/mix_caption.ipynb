```python
import os 
import json
from collections import Counter
```

```python
# 말투가 바뀌지 않은 원본 캡션의 갯수 (이미지 1개 당)
LEFT_NUM = 4

TITLE = '8_2'

#가져온 이미지의 경로 2만개
TO_DATA_PATH = '../data/pbl_raw/pbl_train_raw_data_20000_100051/'

with open('../data/pbl_raw/annotations/pbl_train_raw_caption_20000_100051.json', 'r') as f:
    before_caption = json.load(f)

with open('../data/pbl_change/annotations/pbl_train_change_caption_20000_100051_100051.json', 'r') as f:
    after_caption = json.load(f)
```

```python
image_ids = [int(item['image_id']) for item in before_caption]
image_id_counts = Counter(image_ids)
image_id_counts = dict(image_id_counts)
for key in image_id_counts:
    image_id_counts[key] -= LEFT_NUM
image_id_counts
```

```python
#image_id 전처리
title = os.listdir(TO_DATA_PATH)
title = [t.split('.')[0] for t in title] #remove .jpg
title = [t.split('_')[-1] for t in title] #get 0000000000009
title = [int(t) for t in title] #get 9

#키로 사용할 image id
left = {keys : LEFT_NUM for keys in title}
right = image_id_counts

print(len(title))
print('----------------------------------------------------------------------------------------------------')
print(title)
```

```python
res = []
#바뀌기 전 캡션
#10만개 캡션 중 모든 image_id에 대해 LEFT_NUM개 캡션만 골라내기

#caption[:10]
print(len(before_caption))
print(before_caption[:10])

#걸러지는 캡션 개수
left_flag = 0
#left에 해당하는 캡션만 골라내기
for c in before_caption:
    img_id = int(c['image_id'])

    if img_id not in left.keys():
        continue

    if left[img_id] > 0:
        left[img_id] -= 1
        res.append(c)
```

```python
#LEFT_NUM 캡션 개수
print(len(res))
```

```python
#바뀐 후 캡션

right_flag = 0
for c in after_caption:
    img_id = int(c['image_id'])

    if img_id not in right.keys():
        continue

    if right[img_id] > 0:
        right[img_id] -= 1
        res.append(c)
```

```python
print(len(res))
print(left_flag, right_flag)

print(res[:10])
print('-------------------------------------------------------')
print(res[-10:])
```

```python
#json 파일로 저장
with open('../data/pbl_mix/annotations/pbl_train_mix_caption_'+TITLE+'_20000_100051.json', 'w') as f:
    json.dump(res, f)
```
