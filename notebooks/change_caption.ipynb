```python
from openai import OpenAI
import json

client = OpenAI()
```

```python
########################마지막 결과를 로드해야함!########################
with open('../data/pbl_change/annotations/pbl_train_change_caption_20000_100051_53000.json', 'r') as f:
    caption_list = json.load(f)

#caption[:10]
print(caption_list[:10])
```

```python
sysPrompt = """
You are sentence tone converter.
Change the tone to a teenager.
The output must be a single sentence, not multiple sentences.

Example : 
Before : A long haired cat sleeps on the hood of a car.
After : Super cute cat with long hair is totally crashed out on the hood of a car.
"""
```

```python
def change_caption(caption):
    
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": sysPrompt},
            {"role": "user", "content": caption}
        ],
        # max_tokens = 20
        max_tokens = 50
    )
    
    return completion.choices[0].message.content
```

```python
########################시작 인덱스 조정 필수!########################
start_index = 53000
end_index = len(caption_list)

idx = start_index

for item in caption_list[start_index:end_index]:
    print("current idx : " + str(idx))
    
    if idx % 100 == 0:
        print('Before : ' + item['caption'])
        with open(f'../data/pbl_change/annotations/pbl_train_change_caption_20000_100051_{idx}.json', 'w') as f:
            json.dump(caption_list, f)
            
    chg_caption = change_caption(item['caption'])
    item['caption'] = chg_caption
    
    if idx % 100 == 0:
        print('After : ' + item['caption'])
        
    idx += 1
```

```python
with open(f'../data/pbl_change/annotations/pbl_train_change_caption_20000_100051_{idx}.json', 'w') as f:
    json.dump(caption_list, f)
```
